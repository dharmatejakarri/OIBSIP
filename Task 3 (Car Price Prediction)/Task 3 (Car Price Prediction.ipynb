```python

plt.xlabel('Name of the car')
# naming the y axis
plt.ylabel('Price of the Car')
  
# giving a title to my graph
plt.title('Car price prediction Graph')
plt.bar(data['CarName'], data['price'],
        width = 0.8, color = ['red', 'green'])
  
# function to show the plot
plt.show()
```
